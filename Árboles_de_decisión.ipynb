Cargamos las librerias de trabajo

```python
import pandas as pd
import numpy as np

#librerias arbol
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import confusion_matrix

#conectar al drive -> vamos a tener los archivos disponibles
from google.colab import drive
drive.mount('/content/drive')
```

```python
nxl='/content/drive/MyDrive/ANALÍTICA DE NEGOCIOS /6. SolicitantesCrédito(USD) (3).xlsx' #NXL SIGNIFICA NAME EXCEL
XDB=pd.read_excel(nxl, sheet_name=1) #base AAA
XDB=XDB.dropna()
XDB.head(100)



```

**Descripción de los datos**


De acuerdo con la bd de solicitantes de crédito, esta posee un total de 5842 datos totales divididos en 2959 prenegados y 2883 preaprobados. Cada uno de los solicitantes estará descrito en función de 4 variables,
* Edad: Indica el número de años de la persona
* Ingresos: Indica la cantidad de ingresos totales registrados en un mes por el solicitante  
* Egresos: Indica los gastos en los que incurre en el mes una persona pra cubirir sus necesidades básicas
* Monto (EAD): Es la cantidad solicitada de una persona a la fintech
* PreApr: Indica la decisión del banco frente al otorgamiento del crédito

```python
#seleccionar variables
XD=XDB[['Edad','Ingresos','Egresos','Monto (EAD)']]
YD=XDB[['PreApr']]
```

2. se procede con la implememntación del modelo

```python
mar=DecisionTreeClassifier(criterion='gini',max_depth=4)
mar.fit(XD,YD)

#se procede con la gráfica
from pydotplus import graph_from_dot_data #me crea la gráfica
from sklearn.tree import export_graphviz #me exporta a png, jpg, tiff

ve= ['Edad','Ingresos','Egresos','Monto (EAD)']
dot_data=export_graphviz(mar,feature_names=ve)
graph_from_dot_data(dot_data).write_png('2. ArbolDeDecisión_M.png')


```

3. se procede con el pronóstico de los solicitantes que no poseen preaprobación

```python
XDB0=pd.read_excel(nxl, sheet_name=0)
XDB0.head(100)
#vamos a clasificar los que están en la hoja 1 con los solicitantes de crédito de la hoja 2
#Se cre al bd a pronosticar
XD0=XDB0[['Edad','Ingresos','Egresos','Monto (EAD)']]
YD0=XDB0[['PreApr']]

#determinar filas de solicitantes que no poseen preaprb
filas=np.where(np.isnan(YD0))[0]
XD0=np.array(XD0)
XD0=XD0[filas]

#le calculaos la preaprobación a los solicitantes
y_pred=mar.predict(XD0)
print(y_pred)

#se crea el informe en excel
df=pd.DataFrame(np.column_stack((filas, XD0,y_pred)))
#filas donde estan los solicitantes que no poseen preapr, XD0 los datos de esas eprsonas, y_pred la decisión
df.columns=['Solicitante', 'Edad','Ingresos','Egresos','Monto (EAD)','PreApr']
df.to_excel("Evaluación créditos.xlsx")
```

4. Evaluar el comportamiento del modelo

```python
ydp=mar.predict(XD) #pronóstico sobre la bd original
cm=confusion_matrix(YD,ydp)
print("La matriz de confusión es:\n",cm)
VN=cm[0,0];FP=cm[0,1];FN=cm[1,0];VP=cm[1,1]


#Métricas de desempeño
EX=(VP+VN)/(VP+VN+FP+FN) #Exactitud
print('La exactitud del modelo es:\n',EX)

TE=(FN+FP)/(VP+VN+FP+FN) #Tasa de error:
print('La tasa de error del modelo es:\n',TE)

SE=VP/(VP+FN) #Sensibilidad
print('La sensibilidad del modelo es:\n',SE)

SP=VN/(VN+FP) #Especificidad
print('La especificidad del modelo es:\n',SP)

PR=VP/(VP+FP) #Precisión
print('La precisión del modelo es:\n',PR)

PRNeng=VN/(VN+FN) #Precisión Negativa
print('La precisión negativa del modelo es:\n',PRNeng)
```
